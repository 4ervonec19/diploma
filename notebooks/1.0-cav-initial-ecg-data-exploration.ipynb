Начальный взгляд на данные

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from dotenv import load_dotenv
load_dotenv()

import warnings
warnings.filterwarnings('ignore')
````

Данные

````python
first_dataset = pd.read_csv(os.environ['ECG_DATA_PATH_FIRST'], delimiter='\t', encoding='cp1251')
first_dataset.head()
````

````python
second_dataset = pd.read_csv(os.environ['ECG_DATA_PATH_SECOND'], delimiter='\t', encoding='cp1251')
second_dataset.head()
````

````python
third_dataset = pd.read_csv(os.environ['ECG_DATA_PATH_THIRD'], delimiter='\t', encoding='cp1251',
                            	header=None, names=[f'column_{i}' for i in range(16)])
third_dataset.head()
````

````python
fourth_dataset = pd.read_csv(os.environ['ECG_DATA_PATH_FOURTH'], delimiter='\t', encoding='cp1251', 
                             header=None, names=[f'column_{i}' for i in range(16)])
fourth_dataset.head()
````

Имеем 4 датасета с записями сигналов ```ЭКГ``` и ```Реограмм```

Выведем графики этих сигналов

````python
def plot_signals(signals_dataset, label, column_name, lower_time_limit = 0, upper_time_limit = 1000, color = 'red'):
	plt.figure(figsize=(12, 6))
	plt.plot(signals_dataset[column_name], label=label, color=color, linewidth=1.5)
	plt.title(f'Signal in time ({label})', fontsize=14)
	plt.xlabel('Time (discrete)', fontsize=12)
	plt.ylabel('Voltage ($\mu V$)', fontsize=12)
	plt.legend(fontsize=10)
	plt.grid(True, linestyle=':', alpha=0.7)
	plt.xlim
	plt.xlim([lower_time_limit, upper_time_limit])
	plt.show()

plot_signals(signals_dataset=first_dataset, 
			label='ecg No. 1',
			column_name='ЭКГ(16), мкВ', upper_time_limit=3000)
````

Второй датасет

````python
plot_signals(signals_dataset=second_dataset, 
			label='ecg No. 2',
			column_name='ЭКГ(16), мкВ', upper_time_limit=3000, color='blue')
````

Третий датасет

````python
plot_signals(signals_dataset=third_dataset, 
			label='ecg No. 3',
			column_name='column_6', upper_time_limit=3000, color='green')
````

````python
plot_signals(signals_dataset=fourth_dataset, 
			label='ecg No. 4',
			column_name='column_6', upper_time_limit=3000, color='black')
````

Итак, имеется 4 записи сигналов ```ЭКГ.``` Посмотрим на общее число точек в записях

````python
num_first_record = first_dataset['ЭКГ(16), мкВ'].notna().sum()
num_second_record = second_dataset['ЭКГ(16), мкВ'].notna().sum()
num_third_record = third_dataset['column_6'].notna().sum()
num_fourth_record = fourth_dataset['column_6'].notna().sum()

print(f"Число дискретных отсчётов сигнала 1 (ЭКГ): {num_first_record}")
print(f"Число дискретных отсчётов сигнала 2 (ЭКГ): {num_second_record}")
print(f"Число дискретных отсчётов сигнала 3 (ЭКГ): {num_third_record}")
print(f"Число дискретных отсчётов сигнала 4 (ЭКГ): {num_fourth_record}")
print(f"Общее число отсчетов во всех записях (ЭКГ): {num_fourth_record + num_first_record + num_second_record + num_third_record}")
````

Также проверим общее число точек

````python
num_first_record = first_dataset['ТРКГ(32), мОм'].notna().sum()
num_second_record = second_dataset['ТРКГ(32), мОм'].notna().sum()
num_third_record = third_dataset['column_14'].notna().sum()
num_fourth_record = fourth_dataset['column_14'].notna().sum()

print(f"Число дискретных отсчётов сигнала 1 (РЕО): {num_first_record}")
print(f"Число дискретных отсчётов сигнала 2 (РЕО): {num_second_record}")
print(f"Число дискретных отсчётов сигнала 3 (РЕО): {num_third_record}")
print(f"Число дискретных отсчётов сигнала 4 (РЕО): {num_fourth_record}")
print(f"Общее число отсчетов во всех записях (РЕО): {num_fourth_record + num_first_record + num_second_record + num_third_record}")
````

Итак, в распоряжении имеется порядка ```миллиона``` точек

Посмотрим на совместный график ```ЭКГ + Рео```

````python
def plot_signals_common(signals_dataset, label_ecg, 
				 label_reo, 
				 column_name_ecg, 
				 column_name_reo,
				 lower_time_limit = 0, 
				 upper_time_limit = 1000, 
				 color_ecg = 'red',
				 color_reo='blue'):
	plt.figure(figsize=(12, 6))
	plt.plot(signals_dataset[column_name_ecg], label=label_ecg, color=color_ecg, linewidth=1.5)
	plt.plot(signals_dataset[column_name_reo], label=label_reo, color=color_reo, linewidth=1.5)
	plt.title(f'Signals in time ({label_ecg} + {label_reo})', fontsize=14)
	plt.xlabel('Time (discrete)', fontsize=12)
	plt.ylabel('Value', fontsize=12)
	plt.legend(fontsize=10)
	plt.grid(True, linestyle=':', alpha=0.7)
	plt.xlim
	plt.xlim([lower_time_limit, upper_time_limit])
	plt.show()

plot_signals_common(signals_dataset=first_dataset, 
			label_ecg='ecg No. 1',
			label_reo='reo No. 1',
			column_name_ecg='ЭКГ(16), мкВ', 
			column_name_reo='ТРКГ(32), мОм', 
			upper_time_limit=10000)
````

Вторая пара

````python
plot_signals_common(signals_dataset=second_dataset, 
			label_ecg='ecg No. 2',
			label_reo='reo No. 2',
			column_name_ecg='ЭКГ(16), мкВ', 
			column_name_reo='ТРКГ(32), мОм', 
			upper_time_limit=10000)
````

Третья пара

````python
plot_signals_common(signals_dataset=third_dataset, 
			label_ecg='ecg No. 3',
			label_reo='reo No. 3',
			column_name_ecg='column_6', 
			column_name_reo='column_14', 
			upper_time_limit=10000)
````

И четвертая

````python
plot_signals_common(signals_dataset=fourth_dataset, 
			label_ecg='ecg No. 4',
			label_reo='reo No. 4',
			column_name_ecg='column_6', 
			column_name_reo='column_14', 
			upper_time_limit=10000)
````
